````python
# Prompt Template

import yaml

def build_reverse_prompt(row):
    """
    Build the instruction to the LLM:
    - Provide the workflow content
    - Ask the LLM to generate a natural-language prompt that a developer
      would realistically write to request that workflow.
    """

    try:
        content = row.get("workflow_content_file", "")
    except:
        content = ""

    try:
        repo = row.get("repository", "")
    except:
        repo = ""

    try:
        language = row.get("language", None)
    except:
        language = None

    print(row)

    # Parse YAML safely
    try:
        data = yaml.safe_load(content) if content else {}
    except:
        data = {}

    try: 
        wf_name = data.get("name", "CI Workflow")
    except:
        wf_name = "CI Workflow"

    return f"""
    You are given a real GitHub Actions workflow from the repository **{repo}**.

    Your task is to REVERSE ENGINEER a natural-language prompt that a developer would have typed into an LLM to generate this workflow.

    ### Instructions
    - Do NOT output YAML.
    - Do NOT describe the YAML structure.
    - Do NOT include meta/instructional language like this workflow does X.
    - Instead, write a clean, natural-sounding prompt that a developer would actually send to an AI model.
    - The prompt should describe the purpose, triggers, environment, steps, and behavior needed to recreate this workflow accurately.
    - The prompt should feel human, not machine-generated.
    - It should be something a user would type into ChatGPT or Gemini.

    ### Here is the workflow to reverse-engineer:

    ```yaml
    {content}
    ```

    Now, generate the natural-language prompt a human would give to request this workflow.
    """
````

````python
import subprocess, tempfile, json, os

def run_actionlint_on_string(yaml_text: str, timeout=8):
    with tempfile.NamedTemporaryFile(mode="w", suffix=".yml", delete=False, encoding="utf-8") as f:
        f.write(yaml_text or "")
        temp_path = f.name

    try:
        try:
            result = subprocess.run(
                ["actionlint", "-format", "{{json .}}", temp_path],
                capture_output=True,
                text=True,
                shell=False,
                timeout=timeout
            )
        except subprocess.TimeoutExpired:
            return {"exit_code": -1, "diagnostics": [], "raw_output": None, "stderr": "TIMEOUT"}

        stdout = (result.stdout or "").strip()
        stderr = (result.stderr or "").strip()

        diagnostics, raw_output = [], None
        if stdout:
            try:
                diagnostics = json.loads(stdout)
            except json.JSONDecodeError:
                raw_output = stdout

        return {"exit_code": result.returncode, "diagnostics": diagnostics, "raw_output": raw_output, "stderr": stderr}
    finally:
        os.remove(temp_path)
````

````python
def build_repair_prompt(nl_prompt: str, prev_yaml: str, lint: dict) -> str:
    diagnostics = lint["diagnostics"] if lint.get("diagnostics") else lint.get("raw_output") or lint.get("stderr") or ""
    diagnostic_text = json.dumps(diagnostics, ensure_ascii=False, indent=2) if isinstance(diagnostics, (list, dict)) else str(diagnostics)

    return f"""
        You are repairing a GitHub Actions workflow YAML to satisfy actionlint.

        Original user request (do NOT change intent):
        {nl_prompt}

        Previous YAML (has errors):
        {prev_yaml}

        actionlint diagnostics (authoritative):
        {diagnostic_text}

        Repair requirements:
        - Output ONLY valid GitHub Actions YAML (no markdown fences, no explanations).
        - Make the smallest edits necessary to fix the diagnostics.
        - Preserve the workflow's purpose, triggers, and overall behavior.
        - Do not invent repo-specific scripts/paths; if a step command is clearly invalid syntactically, fix syntax only.
        """
````

````python
import re

def normalize_yaml(text: str) -> str:
    if not text:
        return ""
    t = text.strip()
    t = re.sub(r"^```(?:yaml)?\s*", "", t, flags=re.IGNORECASE)
    t = re.sub(r"\s*```$", "", t)
    return t.strip() + "\n"
````

````python
# THIS ONE IS FOR GEMINI 2.5 FLASH - THIS IS USING THE FLASH GENERATED WORKFLOWS AS PASS 0
# REWRITE THIS LATER FOR 2.5 PRO AND SWITCH THE GENERATED_WORKFLOW COLUMN FETCH TO "generated_workflow"
# GENERATED_WORKFLOW_MODEL = "generated_workflow_2.5_pro"
GENERATED_WORKFLOW_MODEL = "generated_workflow_2.5_flash"


import time
import pandas as pd
from tqdm import tqdm

K = 5  # max repair iterations (Pass@K)
MODEL_NAME = "gemini-2.5-flash"  # start cheap; you can swap later

def generate_yaml(model_fn, prompt: str) -> str:
    return normalize_yaml(model_fn(prompt) or "")

# model_fn should be your Gemini call that returns response.text
# e.g., model_fn = generate_yaml_solutions

for index, row in tqdm(df_rq3.iterrows(), total=df_rq3.shape[0]):

    if index in df_rq3_iterative_results["index"].unique():
        # print(f"Skipping index {index}, already processed.")
        continue  # Skip already processed

    nl_prompt = df_rq3.at[index, "generated_prompt"]

    # Start from existing generated_workflow if present, else generate fresh
    cur_yaml = df_rq3.at[index, GENERATED_WORKFLOW_MODEL]
    if pd.isna(cur_yaml) or not str(cur_yaml).strip():
        cur_yaml = generate_yaml(generate_yaml_solutions, nl_prompt)
    else:
        cur_yaml = normalize_yaml(str(cur_yaml))

    history = []
    for it in range(1, K + 1):

        # print(f"Index {index} - Repair Iteration {it}")

        t0 = time.time()
        lint = run_actionlint_on_string(cur_yaml)
        dt = time.time() - t0

        with open("repair_iteration_logs_intermediate_yaml_workflows.jsonl", "a", encoding="utf-8") as f:
            record = {
                "index": index,
                "iteration": it,
                "yaml": cur_yaml,
                "lint_result": lint,
                "lint_time_sec": dt
            }
            f.write(json.dumps(record, ensure_ascii=False) + "\n")

        history.append({
            "iter": it,
            "exit_code": lint["exit_code"],
            "n_diag": len(lint["diagnostics"]) if lint["diagnostics"] else 0,
            "has_raw": bool(lint["raw_output"]),
            "lint_time_sec": dt,
        })

        if lint["exit_code"] == 0:
            break  # success

        repair_prompt = build_repair_prompt(nl_prompt, cur_yaml, lint)
        cur_yaml = generate_yaml(generate_yaml_solutions_2_5_flash, repair_prompt)

    df_rq3.at[index, "rq3_final_yaml"] = cur_yaml
    df_rq3.at[index, "rq3_history"] = json.dumps(history, ensure_ascii=False)
    df_rq3.at[index, "rq3_pass_at_k"] = (run_actionlint_on_string(cur_yaml)["exit_code"] == 0)
    df_rq3.at[index, "rq3_iters_used"] = len(history)
````
